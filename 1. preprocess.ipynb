```python
'''상관계수가 높은 변수들이 매우 많지만, 변수의 의미를 알 수 없다는 한계 존재
중요 변수의 무차별적 제거를 방지하기 위해 상관계수가 
0.9 이상 혹은 -0.9 이하인 변수들만을 정제

Since all columns given are encrypted, there is clear limitation to which one
could interpret the significances of these columns.
In order to refrain from random removal of essential columns, remove columns that 
only show correlation values above 0.9 or below -0.9 (both inclusive)'''

#상관계수가 -0.9 이하인 경우는 전무해 0.9 이상인 것만 고려
'''Take into account for correlation values above 0.9 (inclusive) since there are no
cases in which that has below -0.9 for correlation value'''
np.sum([cor <= -0.9]) 
```

```python
def findCorrelation(df, thresh):
    corr = df.corr(method = 'pearson')
    #제거 대상 변수명
    #'col_drop' variable to contain column names subjected to removal
    col_drop = set() 
    
    for i in range(len(corr.columns)):
        for j in range(i):
            #상관계수가 threshold(0.9) 이상이며 아직 제거 대상에 속해있지 않을 경우
            '''cases when correlation value is above threshold of 0.9 and 
            is yet contained in 'col_drop'''
            if (corr.iloc[i,j] >= thresh) and (corr.columns[j] not in col_drop): 
                colname = corr.columns[i]
                col_drop.add(colname)
                if colname in df.columns:
                    #해당 column 삭제
                    #remove columns that match condition 
                    del df[colname] 
    
    return df, col_drop

findCorrelation(data, 0.9) 
#column 개수 163개 -> 124개로 정제 
#number of columns has been reduced to 124 (prior: 163)
```

### d. 결측치 처리: MICE
### d. NA Imputation : MICE

```python
# 매우 많은 변수들에서 missing이 발생한 데이터셋의 현 상황을 고려해 MICE 선택
# MICE approach appropriate for our circumstance, in which several columns show NA
columns = data.columns

imputer = IterativeImputer(random_state=26)
mice = imputer.fit(data)

data = pd.DataFrame(mice.transform(data))

#column명 변경
#Change names of columns
data.columns = columns  

#MICE 후 결측치 유무 확인
#Check if there are NA values after MICE transformation
data.isnull().values.any() 
```

### e. 클래스 불균형 처리 : SMOTE
### e. Dealing with imbalanced datasets: SMOTE

```python
# over-sampling 전 데이터의 분포 비교
# check data distribution of the column 'class' before over-sampling
plt.bar(np.arange(2), data['class'].value_counts(), tick_label = ['neg','pos']) 
```

```python
y = data['class']
x = data.drop('class', axis = 1)

#SMOTE를 이용한 over-sampling
#oversamplling using SMOTE
smote = SMOTE(random_state = 26).fit_sample(x, y) 
```

```python
#over-sampling된 x를 데이터프레임 형태로 변환
#Change over-sampled x values into dataframe
resample_x = pd.DataFrame(smote[0])

#over-sampling된 y를 데이터프레임 형태로 변환
#Change over-sampled y values into dataframe
resample_y = pd.DataFrame(smote[1]) 

#하나의 데이터프레임으로 합치기
#concatenate into a single dataframe
resample_y.columns = ['class']
data = pd.concat([resample_x, resample_y], axis = 1) 

# over-sampling 후 데이터의 분포 비교
# check data distribution of the column 'class' post over-sampling
plt.bar(np.arange(2), data['class'].value_counts(), tick_label = ['neg','pos'])
```

### f. 차원 축소 : 분산을 이용한 변수 선택, PCA
### f. Dimensionality Reduction: Feature selection using data variance, Principal Component Analysis (PCA)

#### (1) 분산이 0인 변수 삭제
#### (1) Remove column with variance value of 0

```python
# 분산이 0인 변수 : 모든 관찰값이 같은 값을 가지므로 무의미
# Column with variance value of 0 : meaningless, as all observations show same values
y = data['class']
x = data.drop('class', axis = 1)

# 기준값 만들기
# set threshold value
thresholder = VarianceThreshold(threshold = 0)

#기준값보다 높은 특성 선택
#Choose features that show values above threshold
var_drop = thresholder.fit(x) 

# test set 전처리를 위한 모델 저장
# save model for later purposes
joblib.dump(var_drop, 'var_drop.pkl')

x = thresholder.transform(x)
x = pd.DataFrame(x)

#1개의 변수 삭제 확인
#Removal of 1 column, making it 122 (prior:123)
x.shape 
```

#### (2) PCA

```python
# 표준화
# Standardization using StandardScaler
scale = StandardScaler().fit(x) 

# test set 전처리를 위한 모델 저장
# save model for future purposes
joblib.dump(scale, 'standard.pkl')

scaled_x = scale.transform(x)

# PCA
# 95%의 분산 설명을 목표로 차원 축소
# reduce dimensionality as to retaining 95% variance of all datasets
pca = PCA(n_components=0.95) 
PCA = pca.fit(scaled_x)

# test set 전처리를 위한 모델 저장
# save model for future purposes
joblib.dump(PCA, 'pca.pkl')

features_pca = PCA.transform(scaled_x)

# 결과 확인
# Check results, reduced to 54 (prior:122)
print("원본 특성 개수:", scaled_x.shape[1])
print("줄어든 특성 개수:", features_pca.shape[1])
```

```python
# 각 PC가 설명하는 분산 시각화
# Visualization of the variance each PC retains
PC = range(1, pca.n_components_+1)
plt.bar(PC, pca.explained_variance_ratio_, color='gold')
plt.xlabel('Principal Components')
plt.ylabel('Variance %')
plt.xticks(PC, rotation = 90, fontsize = 5)
```

```python
#각 PC가 설명하는 분산
#Variance that each PC retains
pca.explained_variance_ 
```

```python
##PCA 결과를 데이터프레임 형태로 변환
##Change PCA results to dataframe
data = pd.DataFrame(data = features_pca) 
data = pd.concat([data, y], axis=1)

data.head()
```
